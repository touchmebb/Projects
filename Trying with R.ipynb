# What is MCMC?

The Markov Chain Monte Carlo (MCMC) method is an algorithm to simple from a distribution. Here is a question, if the distribution is known, why would we need to sample from it?

Well, turns out sampling from a distribution is one of the simpliest way of solving some problems. MCMC is used to draw samples form the posterior probability distribution of some model in Bayesian inference. From there, we can have a glimpse of the parameters we are looking for (e.g. the mean $\mu$ and the variance $\sigma^2$).

```r
set.seed(1)
samples <- rnorm(10000, 0, 1)
```
